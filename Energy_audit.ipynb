```python
import matplotlib.pyplot as plt

# Categories and scores
categories = ["Body", "Mind", "Emotions", "Spirit"]
scores = [4, 1, 2, 2]
colors = ["red", "green", "orange", "orange"]

# Creating the bar chart
plt.figure(figsize=(6, 4))
plt.barh(categories, scores, color=colors)

# Adding labels and title
plt.xlabel("Score")
plt.title("Energy Management Scores")
plt.xlim(0, 5)
plt.gca().invert_yaxis()  # Invert y-axis for better readability

# Display the chart
plt.show()
```

# Energy Management Scores

- **Body (4) 🔴** – A full-fledged energy crisis  
- **Mind (1) 🟢** – Strong energy management skills  
- **Emotions (2) 🟠** – Significant deficits  
- **Spirit (2) 🟠** – Significant deficits  

## How to Evaluate Your Scores:
- **0 (🟩 Excellent):** You have outstanding energy management skills in this area. Keep up the great work!  
- **1 (🟩 Strong):** You manage energy well here, but there’s still room for improvement.  
- **2 (🟠 Deficits):** You may struggle with energy in this area. Consider strategies to improve balance.  
- **3 (🔴 Poor):** Energy management is weak, and this may be affecting your overall well-being.  
- **4 (🔴 Crisis):** This area is in urgent need of attention to avoid burnout or major setbacks.
